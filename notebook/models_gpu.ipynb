```python
%%bash
cd /content/yelp-prediction
mkdir -p data/preds/logs

for m in resnet18 resnet50 resnet152 vgg19 vit_b_16; do
  echo "=== TRAIN $m ==="
  poetry run python src/yelp_prediction/main.py \
    --model $m --epochs 20 --batch-size 128 --lr 0.001 --split-seed 42 \
  | tee "data/preds/logs/train_${m}.log"
done
```

```python
import re
from pathlib import Path
import pandas as pd

log_dir = Path("/content/yelp-prediction/data/preds/logs")
pattern = re.compile(
    r"Epoch\s+(\d+)\s+\|\s+Train Loss:\s+([0-9.]+)\s+\|\s+Val MAE:\s+([0-9.]+)\s+\|\s+Val RMSE:\s+([0-9.]+)"
)

rows = []
for log_path in sorted(log_dir.glob("train_*.log")):
    model = log_path.stem.replace("train_", "")
    text = log_path.read_text(errors="ignore")
    for m in pattern.finditer(text):
        epoch = int(m.group(1))
        train_loss = float(m.group(2))
        val_mae = float(m.group(3))
        val_rmse = float(m.group(4))
        rows.append({
            "model": model,
            "epoch": epoch,
            "train_loss": train_loss,
            "val_mae": val_mae,
            "val_rmse": val_rmse,
        })

df = pd.DataFrame(rows).sort_values(["model", "epoch"])
df.head(), df["model"].unique(), df.shape
```
